```python
#불러온 데이터 미리보기 및 기본 정보 확인
```

```python
#데이터 미리보기
```

```python
#head() 함수는 데이터 프레임의 상위 몇 개의 행을 미리보기함. 기본적으로 5개 행을 보여줌.
import pandas as pd

#CSV 파일 불러오기
df = pd.read_csv('test.csv')

# 상위 5개 행 미리보기
print(df.head())
```

```python
#특정 개수의 행을 미리보기
print(df.head(10))
```

```python
#tail()로 하위 데이터 미리보기
#tail() 함수는 하위 몇 개의 행을 미리보기 함.

print(df.tail())
```

```python
#특정 개수의 하위 행을 보고 싶다면?
print(df.tail(10))
```

```python
#데이터프레임의 기본 정보 확인하기
```

```python
#info()로 데이터프레임의 요약 정보 확인
# info() 함수는 데이터프레임의 전체 구조를 한 눈에 보여줌.
# 각 열의 데이터 타입과 널 값 여부를 확인할 수 있음.
```

```python
print(df.info())
```

```python
#describe()로 주요 통계 정보 확인
#describe() 함수는 숫자형 데이터에 대한 주요 통계 정보를 제공함.
#평균, 표준편차, 최소값, 최대값 등을 확인할 수 있음.
```

```python
print(df.describe())
```

```python
#특정 열에 대한 통계를 보고 싶다면?
print(df['상권번호'].describe())
```

```python
#columns와 index로 열과 행 정보 확인
# 데이터프레임의 열 이름과 행 인덱스를 확인할 수 있음.

print(df.columns)
print("줄 구분")
print(df.index)
```

```python
#데이터프레임의 개별 데이터 롹인하기
```

```python
#loc[]로 헹과 열의 위치를 지정하여 접근하기
#loc[]은 라벨(인덱스, 컬럼명)을 사용해 특정 데이터를 조회.
```

```python
#특정 행 조회
print(df.loc[0]) #첫 번째 행 조회
```

```python
#특정 행과 열 조회
print(df.loc[0, '상권명']) # 첫 번쨰 행의 '상권명' 열 데이터
```

```python
#iloc[]로 행과 열의 위치 인덱스를 사용해 데이터를 조회.
# iloc[]은 정수 인덱스를 사용해 데이터를 조회.

#첫 번째 행 조회
print(df.iloc[0])
```

```python
#첫 번째 행의 두 번째 열 데이터 조회
print(df.iloc[0,1])
```

```python
#특정 열이나 행 전체를 선택하기
```

```python
#특정 열 전체를 선택
print(df['상권명'])
```

```python
#여러 열을 선택하고 싶다면?
print(df[['상권명','시도명']])
```

```python
#특정 행 전체를 선택하려면 loc[] 또는 iloc[]을 사용할 것.
```
